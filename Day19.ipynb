```python
import numpy as np
from math import atan2, degrees
from intcode_computer import intcode_computer
from sys import stdout
```

```python
cpu = intcode_computer(19)
```

```python
char = ['.','#', 'O']
area = np.zeros((50,50), dtype=np.int)
for y in range(0, 50):
    for x in range(0, 50):
        cpu.reset_memory()
        output = cpu.run_program([x,y],pause_on_out=True,verbose=False)
        area[x,y] = output[0]
        stdout.write(char[output[0]])
    stdout.write('\n')
print('Part 1 Solution:', np.sum(area))
```

```python
#Find left edge of beam and check the diagonal in in the beam
a = 100-1
y = 100
x = 0
while True:
    #scan x
    while True:
        cpu.reset_memory()
        output = cpu.run_program([x,y],pause_on_out=True,verbose=False)[0]
        if output == 1:
            break
        x += 1
    #check diagonal
    cpu.reset_memory()
    output = cpu.run_program([x+a,y-a],pause_on_out=True,verbose=False)[0]
    if output == 1:
        y -= a
        break
    y += 1
    
print('Part 2 Solution:', (10000*x) + y)
```
