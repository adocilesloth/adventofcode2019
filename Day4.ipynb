```python
import numpy as np
```

```python
def reset():
    start = np.array([2,7,3,0,2,5])
    stop =  np.array([7,6,7,2,5,3])
    return start, stop
```

```python
def incriment(value):
    for i in range(len(value)-1, 0, -1):
        if value[i] == 10:
            value[i-1] += 1
            value[i] = 0
    for i in range(0, len(value)-1):
        if value[i] > value[i+1]:
            value[i+1] = value[i]
    return value

def get_end(value):
    for i in range(0, len(value)-1):
        if value[i] > value[i+1]:
            value[i] -= 1
            for j in range(i+1, len(value)):
                value[j] = 9
    return value

def check_pw(value, part=1):
    pairs = np.array([False, False, False, False, False])
    for i in range(0, len(value)-1):
        if value[i] == value[i+1]:
            pairs[i] = True
            
    num_pairs = np.sum(pairs)
    
    #Part 1
    if part == 1:
        if num_pairs > 0:
            return True
        else:
            return False
    
    #Part 2
    #No pairs
    if num_pairs == 0:
        return False
    
    #One pair
    elif num_pairs == 1:
        return True
    
    #Two Pairs
    elif num_pairs == 2:
        #One triple
        for i in range(0, 4):
            if pairs[i] == pairs[i+1] == True:
                return False
        #Two pairs
        return True
    
    #Three pairs
    elif num_pairs == 3:
        #Three pairs
        if pairs[0] == pairs[2] == pairs[4]:
            return True
        #One Quad
        for i in range(0, 3):
            if pairs[i] == pairs[i+1] == pairs[i+2]:
                return False
        #One pair, one trip
        return True
    
    #Four pairs
    elif num_pairs == 4:
        #Two trip
        if pairs[2] == False:
            return False
        
        #One pair one quad
        if pairs[0] == pairs[4] == True:
            return True
        
        #One quint
        return False
        
        
    
    #Five pairs
    elif num_pairs == 5:
        #One sextuple
        return False
    
    return False
```

```python
pw = 0
start, stop = reset()
value = incriment(start)
end = get_end(stop)

print(value, '-', end)

while True:
    if check_pw(value):
        pw += 1
    value[-1] += 1
    
    if value[-1] == 10:
        value = incriment(value)
        
    if (value == end).all():
        if check_pw(value):
            pw += 1
        break
print('Part 1 Solution:', pw)
```
