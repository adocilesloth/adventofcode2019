```python
layouts = [area]
run = True
while run:
    new_area = np.empty((len(area), len(area[0])), dtype=np.str)
    for i in range(0, len(area)):
        for j in range(0, len(area[i])):
            #Count bugs, being careful with edge positions
            bugs = 0
            if i > 0 and area[i-1,j] == '#':
                bugs += 1
            if i < len(area)-1 and area[i+1,j] == '#':
                bugs += 1
            if j > 0 and area[i,j-1] == '#':
                bugs += 1
            if j < len(area[i])-1 and area[i,j+1] == '#':
                bugs += 1

            #If uninfested, check if we can infest
            if area[i,j] == '.':
                if bugs == 1 or bugs == 2:
                    new_area[i,j] = '#'
                else:
                    new_area[i,j] = '.'

            #If infested, check we can survive
            elif area[i,j] == '#':
                if bugs == 1:
                    new_area[i,j] = '#'
                else:
                    new_area[i,j] = '.'
    area = new_area
    for layout in layouts:
        if np.array_equal(layout, area):
            run = False
            data = area.flatten()
            bug = np.where(data == '#')[0]
            biod = 0
            for pwr in bug:
                biod += pow(2, pwr)
            print('Part 1 Solution:', biod)
            break
    layouts.append(area)
```

```python
data = np.genfromtxt('day24_input.txt', dtype=np.str, delimiter='\n', comments=None)

#data = np.array(['....#',
#                '#..#.',
#                '#..##',
#                '..#..',
#                '#....'])

area = []
for line in data:
    area.append(np.array(list(line)))
area = np.array(area)

area[2,2] = '?'

blank = np.array([['.', '.', '.', '.', '.'],
                  ['.', '.', '.', '.', '.'],
                  ['.', '.', '?', '.', '.'],
                  ['.', '.', '.', '.', '.'],
                  ['.', '.', '.', '.', '.']])

volume = {}
volume[0] = area

for h in range(0, 200):
    levels = []
    for key in volume.keys():
        levels.append(int(key))
    
    #Add levels above/below if we will infest them
    mx = np.max(levels)
    if volume[mx][1,2] == '#' or volume[mx][3,2] == '#' or volume[mx][2,4] == '#' or volume[mx][3,2] == '#':
        mx += 1
        volume[mx] = deepcopy(blank)
        levels.append(mx)
    mn = np.min(levels)
    if '#' in volume[mn][0] or '#' in volume[mn][4] or '#' in volume[mn][:,0] or '#' in volume[mn][:,4]:
        mn -= 1
        volume[mn] = deepcopy(blank)
        levels.append(mn)
        
    levels = np.sort(levels)
    
    new_volume = {}
    #Loop over levels
    for level in levels:
        area = volume[level]
        new_area = np.empty((len(area), len(area[0])), dtype=np.str)
        for i in range(0, len(area)):
            for j in range(0, len(area[i])):
                if area[i,j] == '?':
                    new_area[i,j] = '?'
                    continue
                
                bugs = 0
                #Check for bugs in levels above/below
                #Level outside
                if i == 0 and level != mn:
                    if volume[level-1][1,2] == '#':
                        bugs += 1
                elif i == 4 and level != mn:
                    if volume[level-1][3,2] == '#':
                        bugs += 1
                if j == 0 and level != mn:
                    if volume[level-1][2,1] == '#':
                        bugs += 1
                elif j == 4 and level != mn:
                    if volume[level-1][2,3] == '#':
                        bugs += 1
                        
                #Level inside
                if i == 1 and j == 2 and level != mx:
                    for k in range(0, 5):
                        if volume[level+1][0,k] == '#':
                            bugs += 1
                elif i == 2 and j == 1 and level != mx:
                    for k in range(0, 5):
                        if volume[level+1][k,0] == '#':
                            bugs += 1
                elif i == 2 and j == 3 and level != mx:
                    for k in range(0, 5):
                        if volume[level+1][k,4] == '#':
                            bugs += 1
                elif i == 3 and j == 2 and level != mx:
                    for k in range(0, 5):
                        if volume[level+1][4,k] == '#':
                            bugs += 1
                
                #Count bugs, being careful with edge positions
                #Same level:
                if i > 0 and area[i-1,j] == '#':
                    bugs += 1
                if i < len(area)-1 and area[i+1,j] == '#':
                    bugs += 1
                if j > 0 and area[i,j-1] == '#':
                    bugs += 1
                if j < len(area[i])-1 and area[i,j+1] == '#':
                    bugs += 1
                
                #If uninfested, check if we can infest
                if area[i,j] == '.':
                    if bugs == 1 or bugs == 2:
                        new_area[i,j] = '#'
                    else:
                        new_area[i,j] = '.'

                #If infested, check we can survive
                elif area[i,j] == '#':
                    if bugs == 1:
                        new_area[i,j] = '#'
                    else:
                        new_area[i,j] = '.'
                        
                new_volume[level] = deepcopy(new_area)
    
    volume = deepcopy(new_volume)
    
bugs = 0
for level in levels:
    level_bugs = np.where(volume[level] == '#')[0]
    bugs += len(level_bugs)
print('Part 2 Solution:', bugs)
```
